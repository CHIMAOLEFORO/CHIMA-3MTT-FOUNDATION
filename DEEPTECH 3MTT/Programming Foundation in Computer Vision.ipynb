```python
# import necessary libararies like numpy, opencv and matplotlib

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('/content/tim-mossholder-hOF1bWoet_Q-unsplash.jpg')# Reading the image
plt.imshow(img)
```

```python
# image saving

cv.imwrite("new_image.jpg", img) # writes image to file
```

```python
# img grayscaling

from google.colab.patches import cv2_imshow
img = cv.imread('/content/tim-mossholder-hOF1bWoet_Q-unsplash.jpg')
gray_image = cv. cvtColor(img, cv. COLOR_BGR2GRAY)
plt.imshow(gray_image)
cv.imwrite("gray_image.jpg", gray_image) # writes image to file
```

```python
# convert image to rgb
rgb_img = cv. cvtColor(img, cv. COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
# convert image to bgr
bgr_img = cv. cvtColor(img, cv. COLOR_RGB2BGR)
plt.imshow(bgr_img)
```

```python
# image resizing
rgb_img.shape
```

```python
resized_rgb = cv.resize(rgb_img, (4500, 3000)) # downscaling
plt.imshow(resized_rgb)
```
