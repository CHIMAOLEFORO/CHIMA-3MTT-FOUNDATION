###QUESTION 1
1. In Nigeria, one of the booming tech spaces is fintech. How do you think computer vision can
help improve payment?
➢ Biometrics
➢ QR-Code
➢ Swipe and pay.

ANSWER:
Computer Vision Can applied in the following ways to Improve Payments in Fintech Tech Spaces in Nigeria:

➢ Biometrics (Applying Facial Recognition & Fingerprint Authentication Techologies)
•	Fraud Prevention: Facial recognition and fingerprint scanning can serve as better options for authenticating and authorizing transactions in preference to traditional PINs and Passwords, thereby reducing the risk of identity theft, third party intrusion and fraud.
•	Seamless Authentication: Users can authorize transactions simply by looking at their phone or a camera at a payment terminal (Utilizing Facial Recognition Authentication).
•	Contactless Transactions: With face recognition-based payments, customers don’t need to touch any device, improving hygiene and convenience, minimizing the need for numerous gadgets for payment authorization.

➢ QR-Code-Based Payments
•	Automated QR Code Scanning: Integrating Computer vision  can improve the speed and accuracy of scanning QR codes for payments, even in extreme contions of poor/low-light or faint/damaged code conditions.
•	Dynamic QR Codes for Security: AI-enhanced QR codes generation that change dynamically can prevent unauthorized use or fraud.
•	Facial Matching for Double Authentication: Scanning a QR code while simultaneously verifying the payer’s face can add an extra layer of security.(2 Means of Authentication, Multi-Factor Authentication)

➢ Swipe and Pay (Gesture Recognition)
•	Gesture-Based Transactions: Customers can make payments using pre-defined (Pre-Registered) hand gestures or swipes captured by computer vision systems.
•	Card-Free Transactions: Eliminates the need for physical cards; instead, payments can be authorized via hand motion or face ID recognition. It helps the ease of transactions incases where customer cards are unavailable either because they forgot it, misplaced it, damaged and mutilated cards, theft of cards, or delay in issuance of their cards by their financial institutions.
•	Enhanced Accessibility: Helps people with disabilities make payments without needing to enter PINs or handle cash. This feature also allows for the inclusion and Integration of people living with disabilities in society.


### QUESTION 2:In Nigeria, we have healthtech startups combating the free access to healthcare. How do you
think computer vision can help these startups better reach more people to have access to
quality and good healthcare.

ANSWER:
Computer Vision can help Healthtech startups in Nigeria  to make healthcare more accessible in the following ares.

Remote Diagnostics & Telemedicine
•	AI-Powered Symptom Detection: By utilizing Computer vision, systems can scan patient images (skin conditions, eye diseases, etc.), analyze them  and suggest potential diagnoses.
•	Facial Analysis for Early Disease Detection: By utilizing Computer vision symptoms like jaundice, stroke signs, or dehydration can be Detected through facial recognition in video calls or Image scan.

Automated Medical Imaging Analysis
•	Faster Diagnosis: By utilizing Computer vision doctors and other healthcare personnel can read and understand X-rays, MRIs, and CT scans better and quicker thereby reducing diagnosis time.
•	Affordable Healthcare: Startups can deploy AI-based diagnostics in rural areas where doctors are scarce, healthcare facilities are few and unequipped,by helping early detection of diseases prompting patients to quickly seek medical help before the escalation of the ailment.

Smart Pharmacies & Drug Authentication
•	Fake Drug Detection: Scanning medicine packaging using computer vision can verify authenticity, preventing counterfeit drug circulation.
•	Automated Prescription Matching: Leveraging models that have been trained on  handwriting Recognition can be deployed in reading handwritten prescriptions and suggesting accurate medication matches, reducing errors in pharmacies.

Crowd & Patient Flow Management
•	Queue Optimization: AI-powered cameras can analyze hospital or clinic queues, predicting wait times and suggesting better scheduling and helps the Institution management in the review of their policies and better decision making geared towards improved service delivery.
•	Patient Monitoring: Leveraging on Object tracking and segmentation features of Computer Vision, patients in need of urgent care in crowded hospitals can be quickly Identified through real-time video surveillance.

Computer vision in fintech and healthtech has immense potential to improve efficiency, security, and accessibility.

```python
# loading an image
import cv2
from google.colab.patches import cv_imshow
import numpy as np
```

```python
# load image
img = cv2.imread("/content/Corrupted Student Picture.png")
```

```python
# image dimension
img.shape # (height, width, channel)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
# view and image with matplotlib
import matplotlib.pyplot as plt
plt.imshow(img) # View image with matplotlib
```

```python
#VIEWING THE IMAGE WITH CV_IMSHOW

cv_imshow(img) # View image with cv_imshow
```

```python
# VIEWING THE IMAGE IN GRAY SCALE
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv_imshow(gray_image)
```

```python
# Compute Brightness (Mean Pixel Intensity)
brightness = np.mean(img)

# Compute Contrast (Standard Deviation of Pixel Intensities)
contrast = np.std(img)

# Compute Min and Max Intensity Values
min_intensity = np.min(img)
max_intensity = np.max(img)

# Compute Histogram for Distribution
histogram = cv2.calcHist([img], [0], None, [256], [0, 256])

# Print Results
print(f"Brightness: {brightness:.2f}")
print(f"Contrast: {contrast:.2f}")
print(f"Min Intensity: {min_intensity}")
```

```python
# rgb image
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv_imshow(rgb_image)
```

```python
# hsv image
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv_imshow(hsv_image)
```

```python
# contrast / brightness control
# define the alpha and beta
 # Contrast control----2.5 ###Brightness: 137.34 (moderate brightness level)
###Contrast: 64.36 (relatively high contrast)
#Brightness: 142.13
#Contrast: 75.55
alpha = 64.36 # Contrast control----2.5 ###Brightness: 137.34 (moderate brightness level)
beta = 137.34 # Brightness control----20

# call convertScaleAbs function
adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta) # Img, contrast and brightness
cv_imshow(adjusted)
```

```python
# contrast / brightness control
# define the alpha and beta
###value close to original corrupted alpha---3.2, beta---10.01
alpha = 1.2 # Contrast control
beta = 0.1 # Brightness control

# call convertScaleAbs function
adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta) # Img, contrast and brightness
cv_imshow(adjusted)
```

```python
# rgb image
Trgb_image = cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB)
cv_imshow(Trgb_image)
#cntrast alpha---1.2
#Brigtness beta---0.01
```
